```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread("C:\\Users\\saini\\Desktop\\testing\\Edge detection.jpg")
plt.imshow(img)  
```

```python
image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)   #converting image to gray
plt.imshow(image,cmap = 'gray') 
```

```python
#defining lower and upper threshold
lower = 120
upper =240                   
edge_detection = cv2.Canny(image,lower, upper)       #implementing canny algo.
plt.imshow(edge_detection, cmap = "gray")             
```
